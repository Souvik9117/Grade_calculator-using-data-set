```python
import pandas as pd
df=pd.read_csv('students_marks.csv')
print(df)
def calculator_grade(marks):
    if marks>=90:
        return 'A+'
    elif marks>=80:
        return 'A'
    elif marks>=70:
        return 'B'
    elif marks>=60:
        return 'C'
    else:
        return 'D'
df['Total']=df[['Math', 'Science', 'English']].sum(axis=1)
df['Percentage']=df['Total']/3
df['Grade']=df['Percentage'].apply(calculator_grade)
print(df)
```

```python
highest = df['Percentage'].max()
lowest = df['Percentage'].min()


highest_student = df.loc[df['Percentage'] == highest, 'Name'].values[0]
lowest_student = df.loc[df['Percentage'] == lowest, 'Name'].values[0]


print(f"Highest Percentage: {highest} ({highest_student})")
print(f"Lowest Percentage: {lowest} ({lowest_student})")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('students_marks.csv')
def calculator_grade(marks):
    if marks>=90:
        return 'A+'
    elif marks>=80:
        return 'A'
    elif marks>=70:
        return 'B'
    elif marks>=60:
        return 'C'
    else:
        return 'D'
df['Total']=df[['Math', 'Science', 'English']].sum(axis=1)
df['Percentage']=df['Total']/3
df['Grade']=df['Percentage'].apply(calculator_grade)
print(df)
highest=df['Percentage'].max()
lowest=df['Percentage'].min()
average=df['Percentage'].mean()
topper=df.loc[df['Percentage']==highest,'Name'].values[0]
fail_count=df[df['Percentage']<60].shape[0]
print(f'Highest Percentage: {highest} ({topper}) ')
print(f'Lowest Percentage: {lowest}')
print(f'Average Percentage: {average:.2f}')
print(f'Number of Students Failed(<60%):{fail_count}')
df.to_csv('students_report.csv', index=False)
plt.figure(figsize=(4,5))
plt.bar(df['Name'],df['Percentage'], color='skyblue')
plt.xlabel('Student')
plt.ylabel('Percentage')
plt.title('Student Performance')
plt.show()
```

```python
df.to_csv('students_report.csv', index=False)
colors=[]
for p in df['Percentage']:
    if p >= 90:
        colors.append('green')
    elif p >= 80:
        colors.append('blue')
    elif p >= 70:
        colors.append('red')
    else:
        colors.append('purple')

plt.figure(figsize=(4,5))
plt.barh(df['Name'],df['Percentage'], color=colors,edgecolor='black')
plt.xlabel('Percentage')
plt.ylabel('Student')
plt.title('Student Performance')
plt.show()
```

```python
df.to_csv('students_report.csv', index=False)
plt.figure(figsize=(4,5))
plt.hist(df['Percentage'],bins=5,color='salmon',edgecolor='black')
plt.xlabel('Peccentage')
plt.ylabel('Student')
plt.title('Distribution of Student Performance Histogram')
plt.tight_layout()
plt.show()
```
